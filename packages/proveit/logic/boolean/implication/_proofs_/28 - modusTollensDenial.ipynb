```python
from proveit.logic.boolean.implication._theorems_ import modusTollensDenial, modusTollensAffirmation, affirmViaContradiction
from proveit.logic.boolean.negation._theorems_ import fromNegatedFalsification
from proveit.logic.equality._theorems_ import foldNotEquals, notEqualsContradiction
from proveit.logic import Implies, Not, inBool, Equals, NotEquals, Booleans, FALSE
from proveit.common import A, B
```

```python
modusTollensDenial.beginProof([foldNotEquals, modusTollensAffirmation, affirmViaContradiction, fromNegatedFalsification])
```

```python
AeqF = Equals(A, FALSE)
AeqF
```

```python
A.prove([Not(AeqF), inBool(A)])
```

```python
Implies(Not(AeqF), B).prove([Implies(A, B), inBool(A)])
```

```python
AeqF_truth = Implies(Not(AeqF), B).denyingConsequent([Implies(A, B), Not(B), inBool(A), inBool(B)])
AeqF_truth
```

```python
conclusion = AeqF_truth.deriveViaBooleanEquality([Implies(A, B), Not(B), inBool(A), inBool(B)])
conclusion
```

```python
conclusion.generalize(([A], [B]), domains=[Booleans, None], conditions=[Implies(A, B), Not(B)]).qed()
```
