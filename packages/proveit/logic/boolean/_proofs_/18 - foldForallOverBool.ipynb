```python
from proveit.logic.boolean._theorems_ import foldForallOverBool, byCases
from proveit.logic import inBool, Booleans
from proveit.common import A, B, P, PofA
```

```python
foldForallOverBool.beginProof([byCases])
```

```python
conditions = foldForallOverBool.conditions
conditions
```

```python
byCases
```

```python
conclusion = byCases.specialize({B:PofA}, assumptions=[inBool(A)]+list(conditions))
conclusion
```

```python
conclusion.generalize(([P],[A]), domains=[None, Booleans], conditions=conditions).qed()
```
