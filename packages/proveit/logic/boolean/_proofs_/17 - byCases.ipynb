```python
from proveit.logic.boolean._theorems_ import byCases, byCasesWeak
from proveit.logic import inBool, Implies, Not, Equals, NotEquals, Booleans, TRUE
from proveit.common import A, B
```

```python
byCases.beginProof([byCasesWeak])
```

```python
byCasesWeak
```

```python
conditions = byCasesWeak.conditions
conditions
```

```python
BeqT = Equals(B, TRUE)
BeqT
```

```python
Implies(A, BeqT).prove(conditions)
```

```python
Implies(Not(A), BeqT).prove(conditions)
```

```python
byCasesWeak.specialize({A:A, B:BeqT}, assumptions=[inBool(A)]+list(conditions))
```

```python
conclusion = B.prove(list(conditions) + [inBool(A)])
conclusion
```

```python
conclusion.generalize(([A], [B]), domains=[Booleans, None], conditions=conditions).qed()
```

```python
byCases.printRequirements()
```
