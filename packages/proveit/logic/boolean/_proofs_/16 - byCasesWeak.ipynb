```python
from proveit.logic.boolean._theorems_ import byCasesWeak
from proveit.logic.boolean.negation._theorems_ import negationContradiction
from proveit.logic.boolean.implication._theorems_ import modusTollensDenial, affirmViaContradiction
from proveit.logic import inBool, Implies, Not, Booleans
from proveit.common import A, B
```

```python
byCasesWeak.beginProof([negationContradiction, modusTollensDenial, affirmViaContradiction])
```

```python
conditions = byCasesWeak.conditions
conditions
```

```python
Implies(A, B).denyingConsequent([inBool(A), Implies(A, B), Not(B)])
```

```python
B.prove([inBool(A)]+list(conditions) + [Not(B)])
```

```python
Not(B).affirmViaContradiction(B, inBool(A, B)+list(conditions))
```

```python
conclusion = B.prove(inBool(A, B)+list(conditions))
conclusion
```

```python
conclusion.generalize((A, B), domain=Booleans, conditions=conditions).qed()
```
