```python
from proveit.logic.boolean.disjunction._theorems_ import orIfOnlyLeft, trueOrFalse
from proveit.logic.equality._theorems_ import lhsViaEquivalence, rhsViaEquivalence
from proveit.logic import Not
from proveit.common import A, B
```

```python
orIfOnlyLeft.beginProof([trueOrFalse,lhsViaEquivalence,rhsViaEquivalence])
```

```python
conclusion = orIfOnlyLeft.instanceExpr.prove([A, Not(B)])
conclusion
```

```python
conclusion.generalize((A,B), conditions=[A,Not(B)]).qed()
```
