```python
from proveit.logic.boolean.disjunction._theorems_ import orIfOnlyRight, falseOrTrue
from proveit.logic.equality._theorems_ import lhsViaEquivalence, rhsViaEquivalence
from proveit.logic import Not
from proveit.common import A, B
```

```python
orIfOnlyRight.beginProof([falseOrTrue,lhsViaEquivalence,rhsViaEquivalence])
```

```python
conclusion = orIfOnlyRight.instanceExpr.prove([Not(A), B])
conclusion
```

```python
conclusion.generalize((A,B), conditions=[Not(A),B]).qed()
```
