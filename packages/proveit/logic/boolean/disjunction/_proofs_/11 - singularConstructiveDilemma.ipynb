```python
from proveit.logic.boolean.disjunction._theorems_ import singularConstructiveDilemma, binaryOrContradiction
from proveit.logic.boolean.implication._theorems_ import modusTollensDenial, affirmViaContradiction
from proveit.logic import Implies, Or, Not, Equals, inBool, Booleans, TRUE
from proveit.common import A, B, C
from proveit import ExpressionList
```

```python
singularConstructiveDilemma.beginProof([modusTollensDenial, affirmViaContradiction, binaryOrContradiction])
```

```python
conditions = singularConstructiveDilemma.instanceExpr.conditions
conditions
```

```python
Implies(A, Equals(C, TRUE)).denyingConsequent([Implies(A, Equals(C, TRUE)), Not(Equals(C, TRUE)), inBool(A)])
```

```python
Implies(B, Equals(C, TRUE)).denyingConsequent([Implies(B, Equals(C, TRUE)), Not(Equals(C, TRUE)), inBool(B)])
```

```python
assumptions = ExpressionList(list(conditions) + [inBool(A), inBool(B)])
assumptions
```

```python
conclusion = Or(A, B).affirmViaContradiction(Equals(C, TRUE), assumptions)
conclusion
```

```python
conclusion.generalize((A, B, C), domain=Booleans, conditions=conditions).qed()
```
