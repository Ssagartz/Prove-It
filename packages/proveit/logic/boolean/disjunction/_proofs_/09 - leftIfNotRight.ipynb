```python
from proveit.logic.boolean.disjunction._theorems_ import leftIfNotRight, binaryOrContradiction
from proveit.logic.boolean.implication._theorems_ import affirmViaContradiction
from proveit.logic.boolean.negation._theorems_ import contradictionViaNegation
from proveit.logic import Or, Not, inBool, Booleans
from proveit.common import A, B
```

```python
leftIfNotRight.beginProof([binaryOrContradiction, affirmViaContradiction, contradictionViaNegation])
```

```python
conclusion = Or(A, B).affirmViaContradiction(A, [Or(A, B), Not(B), Not(A), inBool(A)])
conclusion
```

```python
conclusion.generalize((A, B), domain=Booleans, conditions=[Or(A, B), Not(B)]).qed()
```
