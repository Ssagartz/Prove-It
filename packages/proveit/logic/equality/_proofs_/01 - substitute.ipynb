```python
from proveit.logic.equality._axioms_ import substitution
from proveit.logic.equality._theorems_ import substitute
from proveit.logic import Equals
from proveit.common import f, P, x, y, Px, Py
```

```python
substitute.beginProof()
```

```python
substitution
```

```python
x_equals_y = Equals(x, y)
x_equals_y
```

```python
Py_equals_Px = substitution.specialize({f:P, x:y, y:x}, assumptions=[x_equals_y])
Py_equals_Px
```

```python
Px_equals_True = Px.evaluate([Px])
Px_equals_True
```

```python
Py_equals_Px.applyTransitivity(Px_equals_True, [Px, x_equals_y])
```

```python
Py_proven = Py.prove([Px, Equals(x, y)])
Py_proven
```

```python
Py_proven.generalize([P,x,y], conditions=[Px, x_equals_y]).qed()
```
