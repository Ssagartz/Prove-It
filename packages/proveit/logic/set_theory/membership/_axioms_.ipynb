```python
from proveit.logic import Forall, And, Not, Equals, InSet, NotInSet, EmptySet
from proveit.number import Exp, Len, Naturals, one
from proveit import Lambda, Iter, Indexed
from proveit._common_ import x, S, f, i, n
%begin axioms
```

**Define an exponentiated set.**

```python
expSetDef = Forall(n, Forall((x, S), Equals(InSet(x, Exp(S, n)), 
                                            And(Equals(Len(x), n), 
                                                Iter(Lambda(i, InSet(Indexed(x, i), S)), one, n)))),
                   domain=Naturals)
```

**$x \notin S$ is defined as the negation of $x \in S$ for all $x$ and $S$:**

```python
notInSetDef = Forall((x, S), Equals(NotInSet(x, S), Not(InSet(x, S))))
```
