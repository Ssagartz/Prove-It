**Define an exponentiated set.**

```python
expSetDef = Forall(n, Forall((xx, S), Equals(InSet(xx, Exp(S, n)), 
                                             And(Equals(Len(xx), n), 
                                                 each_x_in_S))),
                   domain=Naturals)
```

**$x \notin S$ is defined as the negation of $x \in S$ for all $x$ and $S$:**

```python
notInSetDef = Forall((x, S), Equals(NotInSet(x, S), Not(InSet(x, S))))
```
