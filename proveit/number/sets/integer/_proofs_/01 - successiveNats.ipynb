```python
from proveit.number.sets.integer._axioms_ import naturalsDef
from proveit.number.sets.integer._theorems_ import successiveNats
from proveit.logic import InSet
from proveit.common import n, x, S
from proveit.number import Naturals, Add, one
```

```python
successiveNats.beginProof(['proveit.logic'])
```

```python
naturalsDef
```

```python
nInNatsDef = naturalsDef.specialize()
nInNatsDef
```

```python
nInNats = nInNatsDef.lhs
nInNats
```

```python
nInS_hypothesis = nInNatsDef.rhs.instanceExpr.hypothesis
nInS_hypothesis
```

```python
nInS_implication = nInNatsDef.deriveRightViaEquivalence([nInNats]).specialize()
nInS_implication
```

```python
nInS = nInS_implication.deriveConclusion([nInNats, nInS_hypothesis])
nInS
```

```python
nSuccessorInNatsDef = naturalsDef.specialize({n:Add(n, one)})
nSuccessorInNatsDef
```

```python
nSuccessorInS_hypothesis = nSuccessorInNatsDef.rhs.instanceExpr.hypothesis
nSuccessorInS_hypothesis
```

```python
anySuccessorOfSInS = nSuccessorInS_hypothesis.deriveRight([nSuccessorInS_hypothesis])
anySuccessorOfSInS
```

```python
nSuccessorInS = anySuccessorOfSInS.specialize({x:n}, assumptions=[nInNats])
nSuccessorInS
```

```python
nSuccessorInNatsDef_rhs = nSuccessorInS.asImplication(nSuccessorInS_hypothesis).generalize(S)
nSuccessorInNatsDef_rhs
```

```python
nSuccessorInNatsDef.deriveLeftViaEquivalence([nInNats]).generalize(n, domain=Naturals).qed()
```
